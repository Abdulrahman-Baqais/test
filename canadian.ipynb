```python
from sklearn.preprocessing import MinMaxScaler
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM

data['date']=data['date'].astype('category')
data['date'] =data['date'].cat.codes

y_train = data.HC.iloc[:324]
y_test = data.HC.iloc[324:]
train = data.date.iloc[:324]
x_test = data.date.iloc[324:]
train = train[:, np.newaxis]

train = np.reshape(train, (train.shape[0],train.shape[1],1))
train = np.reshape(train, (train.shape[0],train.shape[1],1))
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=True, input_shape=(train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=True, input_shape=(train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=True, input_shape=(train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=True, input_shape=(train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=True, input_shape=(train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=True, input_shape=(train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=True, input_shape=(train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=True, input_shape=(train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=True, input_shape=(train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=True, input_shape=(train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=True, input_shape=(train.shape[1],1)))
model.add(LSTM(units=20))
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(train, y_train, epochs=3, batch_size=1, verbose=2)
```

```python
x_test = x_test[:,np.newaxis]
x_test = np.reshape(x_test, (x_test.shape[0],x_test.shape[1],1))
fore = model.predict(x_test)
```
